```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Part (a)

```python
def cosfunc(amplitude,freq,time,phi):
    return amplitude*np.cos(2*np.pi*freq*time-phi*np.pi/180)

t=np.linspace(0,2,500)
y1=cosfunc(1,10,t,30)

plt.figure(figsize = (10,5))
plt.plot(t,y1)
plt.xlabel('time (s)')
plt.ylabel('y(t)')
plt.title('Cosine wave');
```

Part (b)

```python
y2=cosfunc(0.8,8,t,60)

plt.figure(figsize = (10,5))
plt.plot(t,y1, label='y1')
plt.plot(t,y2, label='y2')
plt.xlabel('time (s)')
plt.ylabel('y(t)')
plt.title('Cosine waves')
plt.legend();
```

Part (c)

```python
y3=y1+y2

plt.figure(figsize = (10,5))
plt.plot(t,y3)
plt.xlabel('time (s)')
plt.ylabel('y(t)')
plt.title('Cosine wave interference');
```

Part (d)
(i)

```python
FFT=np.fft.fft(y3)
print(FFT)
```

Part (d) (ii)

```python
FFT2=np.abs(FFT[:250])
# create a frequency array to be plotted with the Fast Fourier Transform
max_frequency=np.arange(0,125,0.5)
```

Part (d) (iii)

```python
plt.figure(figsize = (5,5))
plt.plot(max_frequency, FFT2)
plt.xlabel('frequency')
plt.ylabel('amplitude')
plt.title('Frequency spectrum of y3')
plt.grid();
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,8)) 
#fig.suptitle('Fourier Transform Sub-plots') 

#ax1.plot.fig(figsize=(5,5))
ax1.plot(max_frequency, FFT2) 
ax1.set_title('Overall Peak') 
ax1.set(xlabel = 'Frequency (Hz)', ylabel = 'Amplitude Transformed') 

ax2.plot(max_frequency, FFT2)
ax2.set_xlim(5, 15)  
ax2.set_title('Peak Zoomed in')
ax2.set(xlabel = 'Frequency (Hz)', ylabel = 'Amplitude Transformed')

plt.tight_layout()

ax1.grid()
ax2.grid();
```
